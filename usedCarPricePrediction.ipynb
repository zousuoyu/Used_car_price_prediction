# 1. Data overview

```python
files = [
    'data/audi.csv', 'data/bmw.csv', 'data/ford.csv', 'data/hyundi.csv', 'data/merc.csv',
    'data/skoda.csv', 'data/toyota.csv', 'data/vauxhall.csv', 'data/vw.csv'
]
dfs = []
for file in files:
    brand = file.split('/')[1].split('.')[0]
    tmp = pd.read_csv(file)
    if brand == 'hyundi':
        tmp = tmp.rename(columns = {'tax(Â£)': 'tax'})
    tmp['brand'] = brand
    dfs.append(tmp)
data = pd.concat(dfs)
print('data size:', data.shape)
data.sample(10)
```

```python
# univariate data explore
for column in data.columns:
    fig = plt.figure(figsize = (16, 6))
    if types.is_object_dtype(data[column].dtype):
        ax = sb.countplot(x = column, data = data)
    else:
        ax = sb.histplot(x = column, data = data, bins = 20)
    for xticker in ax.get_xticklabels():
        xticker.set_rotation(45)
    plt.show()
```

```python
yCol = 'price'
```

```python
def plot(col, yCol, data, idx = 1):
    fig = plt.figure(figsize = (16, 6))
    if idx == 0:
        tmp = data[[col, yCol]].groupby(by = col).agg('mean').reset_index()
        ax = sb.barplot(x = col, y = yCol, data = tmp)
    else:
        ax = sb.scatterplot(x = col, y = yCol, data = data)
    plt.show()
```

```python
for col in data.columns:
    if col != yCol:
        if types.is_object_dtype(data[col].dtype):
            plot(col, yCol, data, 0)
        else:
            plot(col, yCol, data, 1)
```

# 3. Model training and evaluation

```python
# one hot encoder
cateCols = ['brand', 'model', 'fuelType', 'transmission']
df = pd.get_dummies(data, columns = cateCols)
```

```python
def modelTrain(data, model):
    y = data[yCol]
    X = data.drop(columns = [yCol])
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
    
    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    print('train set r2 score:', r2_score(y_train,y_train_pred))
    print('train set rmse:', mean_squared_error(y_train, y_train_pred, squared = False))
    y_test_pred = model.predict(X_test)
    print('test set r2 score:', r2_score(y_test,y_test_pred))
    print('test set rmse:', mean_squared_error(y_test,y_test_pred, squared = False))
```

```python
# ridge regressor
model = Ridge()
modelTrain(df, model)
```

```python
# random foreset regressor
model = RandomForestRegressor()
modelTrain(df, model)
```

```python
# gradient boost regressor
model = GradientBoostingRegressor()
modelTrain(df, model)
```
